ใช้เมื่อไหร่

  ใช้หาข้อมูล เมื่อต้องการค้นหา เช่น เบอร์ อีเมล

```
def creatdatabaselist(n):
  listdb = [] #ประกาศว่าสร้าง list
  for i in range(n): #วนลูป n ครั้ง
    listdb.append(('pan'+str(i),i+100)) #ใส่ข้อมูลเป็น tuple () ใส่ str(n) ใส่ str(n) เป็น element ตัวแรก และ i + 100 เป็น element
  return listdb

n = 1000000 #สร้างข้อมูลจำลอง n record
listdb = creatdatabaselist(n)
listdb
```

```
def creatdatabasedict(n):
  dictdb = {} #ประกาศว่าจะสร้าง dic
  for i in range(n): #วนลูป n ครั้ง
    dictdb['pan'+str(i)] = i + 100 #ใส่ str(n) เป็น key และใส่ i + 100 เป็นค่า
  return dictdb
n = 1000000 #สร้างข้อมูลจำลอง n record
dictdb = creatdatabasedict(n)
dictdb
```

```
def getphonenum(cusname,listdb):
  for c in listdb:
    if c[0]==cusname:
      cnum = c[1]
  return cnum
```

```
getphonenum('pan3',listdb)
```

ทดสอบจับเวลา ในการนำข้อมูลออกมาเทียบ

```
import time

def n_vs_time_list(): #สร้างฟังก์ชัน จับเวลาเทียบกับจำนวนข้อมูลที่เพิ่มขึ้น (ใน list) จาก 10000 ถึง 1000000
  timedata = []
  for n in range(10000,1000000,10000):
    print(n)
    listdb = creatdatabaselist(n)
    starttime = time.time()
    getphonenum('pan9700',listdb)
    endtime = time.time()
    timediff = endtime - starttime
    timedata.append([n,timediff])
  return timedata

timedatalist = n_vs_time_list()
```

```
def n_vs_time_dict(): #สร้างฟังก์ชัน จับเวลาเทียบกับจำนวนข้อมูลที่เพิ่มขึ้น (ใน dict) จาก 10000 ถึง 1000000
  timedata = []
  for n in range(10000,1000000,10000):
    print(n)
    listdb = creatdatabasedict(n)
    starttime = time.time()
    getphonenumdict('pan9700',dictdb)
    endtime = time.time()
    timediff = endtime - starttime
    timedata.append([n,timediff])
  return timedata

timedatadict = n_vs_time_dict()
```
